# **Імпорт бібліотек**

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

1. Зчитати та завантажити тестове чорно-біле та кольорове зображення.

```python
bw_photo = cv2.imread("./bwphoto.jpg", cv2.IMREAD_GRAYSCALE)
color_photo = cv2.imread("./colorphoto.jpg", cv2.COLOR_BGR2RGB)
```


2. Відобразити чорно-біле та кольорове зображення.

```python
plt.figure(figsize=(10, 10))
plt.imshow(bw_photo, cmap='gray')
plt.show();

plt.figure(figsize=(10, 10))
plt.imshow(color_photo)
plt.show();
```

3. Перетворити зображення з колірної моделі BGR в колірні моделі RGB і HSV та відобразити.

```python
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(color_photo, cv2.COLOR_BGR2RGB))
plt.show();

plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(color_photo, cv2.COLOR_BGR2HSV))
plt.show();
```

4. Окремо відобразити колірні канали для зображень із колірними моделями RGB та HSV.

```python
RGBphoto = cv2.cvtColor(color_photo, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(RGBphoto.shape, dtype='uint8')
R[:, :, 0] = RGBphoto[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(RGBphoto.shape, dtype='uint8')
G[:, :, 1] = RGBphoto[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(RGBphoto.shape, dtype='uint8')
B[:, :, 2] = RGBphoto[:, :, 2]
plt.imshow(B)

plt.show()
```

HSV

```python

HSVphoto = cv2.cvtColor(color_photo, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(HSVphoto.shape, dtype='uint8')
R[:, :, 0] = HSVphoto[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(HSVphoto.shape, dtype='uint8')
G[:, :, 1] = HSVphoto[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(HSVphoto.shape, dtype='uint8')
B[:, :, 2] = HSVphoto[:, :, 2]
plt.imshow(B)

plt.show()
```

5. Скопіювати область інтересу (Region of Interest) чорно-білого зображення, використовуючи наступні діапазони 0:img.shape[0]/2, 0:img.shape[1]/2.

```python
crop_bwphoto = bw_photo[0:int(bw_photo.shape[0]/2), 0:int(bw_photo.shape[1]/2)]
print(crop_bwphoto)
plt.imshow(crop_bwphoto, cmap='gray')
plt.show()
```

6. Зберегти область інтересу (Region of Interest).

```python
cv2.imwrite("./save_bw_photo.png", crop_bwphoto)
```
