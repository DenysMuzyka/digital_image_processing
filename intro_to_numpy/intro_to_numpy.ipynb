**1. Створити двомірний нульовий масив розміром 11х11 та заповнити його діагоналі одиницями.**

Імпортуємо бібліотеку NumPy

```python
import numpy as np
```

1.1. Задаємо розмір масиву

```python
size = 11
```

1.2. Створюємо нульовий масив заданого розміру

```python
array = np.zeros([size, size])
```

1.3 Друкуємо створений масив

```python
print(array)
```

1.4 Змінюємо діагоналі за допомогою двох циклів

```python
for k in range(size):
	for n in range(size):
		if (k==n):
			array[k][n] = 1
		elif (n==size-k-1):
			array[n][k] = 1
```

1.5 Виводимо створений масив

```python
print(array)
```
