1.1. Відкрити та завантажити наступні два зображення:
    * assignments/Lab3_half100.png
    * assignments/Lab3_chess100.png

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
lab3_Im_half = cv2.imread("./Lab3_half100.png", cv2.IMREAD_GRAYSCALE)
lab3_Im_chess = cv2.imread("./Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)
```

1.2.Визначити гістограму для кожного зображення:

```python
chess_histogram = cv2.calcHist([lab3_Im_chess],[0],None,[256],[0, 256])
half_histogram = cv2.calcHist([lab3_Im_half],[0],None,[256],[0, 256])
```

1.3.Відобразити завантаженні зображення та відповідно визначені гістограми зображень за допомогою функції plt.subplot:

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(lab3_Im_chess, cmap = 'gray');
plt.subplot(122)
plt.plot(chess_histogram); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(lab3_Im_half, cmap = 'gray');
plt.subplot(122)
plt.plot(half_histogram); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.show();
```
